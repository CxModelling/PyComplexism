```python
from dzdy import *
import matplotlib.pyplot as plt

da = Director()
da.load_pc('scripts/pBAD.txt')
da.load_dc('scripts/BAD.txt')
```

```python
bad = da.new_mc('BAD', 'ABM', tar_pc='pBAD', tar_dc='BAD')
## Key command
bad.add_behaviour('Rein', 'Reincarnation', s_death='Dead', s_birth='Young')
bad.set_observations(states=['Young', 'Middle', 'Old', 'Dead'], transitions=['Die'])
```

```python
mod, out =  da.simulate('BAD', y0={'Young': 10, 'Middle': 10, 'Old': 10}, to=20)
```

```python
out.plot()
plt.show()
```
